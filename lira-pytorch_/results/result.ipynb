```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ['vgg-19', 'vit_large_patch16_224', 'efficientnet_b7']
online_cifar10 = [0.612, 0.516, 0.595]
online_cifar100 = [0.760, 0.606, 0.750]

x = np.arange(len(models))  # label locations
width = 0.35  # bar width

# Plot CIFAR-10 (Online Attack)
fig, ax = plt.subplots()
ax.bar(x, online_cifar10, width, color=['gold', 'orange', 'crimson'], label=models)

ax.set_ylabel('Accuracy')
ax.set_title('Membership Inference Attack (CIFAR-10)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(models)

fig.savefig('./figure/membership_inference_attack_cifar10.png')
plt.show()

# Plot CIFAR-100 (Online Attack)
fig, ax = plt.subplots()
ax.bar(x, online_cifar100, width, color=['gold', 'orange', 'crimson'], label=models)

ax.set_ylabel('Accuracy')
ax.set_title('Membership Inference Attack (CIFAR-100)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(models)

fig.savefig('./figure/membership_inference_attack_cifar100.png')
plt.show()
```
