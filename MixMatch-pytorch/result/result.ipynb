```python
import matplotlib.pyplot as plt
import numpy as np
```

# MixMatch Attack (before defense) 

```python
# Data from the provided image for MixMatch attack
models = ['vgg-19', 'vit_large_patch16_224', 'efficientnet_b7']
n_labeled = ['1K', '4K', '10K']

# Accuracy for CIFAR-10
cifar10_vgg19 = [75.44, 81.97, 81.86]
cifar10_vit = [94.85, 92.1, 97.4]
cifar10_efficientnet = [66.58, 53.32, 60.74]

# Accuracy for CIFAR-100
cifar100_vgg19 = [38.12, 53.25, 61.8]
cifar100_vit = [66.85, 83.85, 87.7]
cifar100_efficientnet = [31, 50.3, 54.75]

x = np.arange(len(n_labeled))  # label locations
width = 0.25  # bar width

fig, ax = plt.subplots()
ax.bar(x - width, cifar10_vgg19, width, label='vgg-19', color='gold')
ax.bar(x, cifar10_vit, width, label='vit_large_patch16_224', color='orange')
ax.bar(x + width, cifar10_efficientnet, width, label='efficientnet_b7', color='crimson')

ax.set_ylabel('Accuracy')
ax.set_title('MixMatch Attack (CIFAR-10)')
ax.set_xticks(x)
ax.set_xticklabels(n_labeled)
ax.legend()

fig.savefig('./figure/mixmatch_cifar10_accuracy.png')

# Plot CIFAR-100 with specific colors
fig, ax = plt.subplots()
ax.bar(x - width, cifar100_vgg19, width, label='vgg-19', color='gold')
ax.bar(x, cifar100_vit, width, label='vit_large_patch16_224', color='orange')
ax.bar(x + width, cifar100_efficientnet, width, label='efficientnet_b7', color='crimson')

ax.set_ylabel('Accuracy')
ax.set_title('MixMatch Attack (CIFAR-100)')
ax.set_xticks(x)
ax.set_xticklabels(n_labeled)
ax.legend()

fig.savefig('./figure/mixmatch_cifar100_accuracy.png')
```

# MixMatch Attack (after defense)

```python

# Data extracted from the provided image
eps = [1, 5, 10, 100]

# Accuracy for CIFAR-10
accuracy_cifar10 = [96.4, 96.42, 96.5, 96.96]
victim_model_cifar10 = 98.4

# Accuracy for CIFAR-100
accuracy_cifar100 = [83.02, 83.14, 83.24, 84.38]
victim_model_cifar100 = 87.86

# Plot for CIFAR-10
fig, ax = plt.subplots()
ax.plot(eps, accuracy_cifar10, marker='o', label='Attacker Model', color='blue')
ax.axhline(y=victim_model_cifar10, color='red', linestyle='--', label='Victim Model')

ax.set_xlabel('Epsilon (eps)')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy vs Epsilon (CIFAR-10)')
ax.set_ylim(94, 99.5)  # Adjusted lower bound
ax.legend()

fig.savefig('./figure/accuracy_vs_epsilon_cifar10.png')
plt.show()

# Plot for CIFAR-100
fig, ax = plt.subplots()
ax.plot(eps, accuracy_cifar100, marker='o', label='Attacker Model', color='blue')
ax.axhline(y=victim_model_cifar100, color='red', linestyle='--', label='Victim Model')

ax.set_xlabel('Epsilon (eps)')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy vs Epsilon (CIFAR-100)')
ax.set_ylim(75, 89.5)  # Adjusted lower bound
ax.legend()

fig.savefig('./figure/accuracy_vs_epsilon_cifar100.png')
plt.show()
```
