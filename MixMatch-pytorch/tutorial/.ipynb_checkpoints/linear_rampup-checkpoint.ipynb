```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
_epochs_1024 = 1024
_epochs_50 = 50

# Linear ramp-up function
def linear_rampup(current, rampup_length):
    if rampup_length == 0:
        return 1.0
    else:
        current = np.clip(current / rampup_length, 0.0, 1.0)
        return float(current)

lambda_u_100 = 100
lambda_u_20 = 20

# Generate data for plotting
values_1024 = [lambda_u_100 * linear_rampup(i, _epochs_1024) for i in range(_epochs_1024)]
values_50 = [lambda_u_20 * linear_rampup(i, _epochs_50) for i in range(_epochs_50)]

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(values_1024, label='Ramp-up (_epochs = 1024)')
plt.plot(range(_epochs_50), values_50, label='Ramp-up (_epochs = 50)', linestyle='--')
plt.title("Linear Ramp-Up Progression")
plt.xlabel("Epoch")
plt.ylabel("Ramp-up (%)")
plt.legend()
plt.grid()
plt.show()
```
