```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Beta distribution
_alpha = 0.75

# Generate l values using Beta distribution
l_values = np.random.beta(_alpha, _alpha, size=10000)

# Plot the distribution of l values
plt.figure(figsize=(8, 5))
plt.hist(l_values, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title(f"Distribution of l (Beta Distribution with alpha={_alpha})")
plt.xlabel("l values")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
